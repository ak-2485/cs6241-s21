## Question 1
Show that if $\phi : \mathcal{V} \rightarrow \mathbb{R}$ is a quadratic form, then $a(v,w) = \phi(v+w)-\phi(v)-\phi(w)$ is a bilinear form for which $\phi(v) = a(v,v)$.


## Answer 1

By definition, $a$ is a bilinear form if 1.-4. hold. 

1. $a(u+v,w) = a(u,w) + a(v,w)$
2. $a(\lambda u,v ) = \lambda a(u,v)$
3. $a(u,v+w) = a(u,v) + a(u,w)$
4. $a(\lambda u,v ) = \lambda a(u,v)$

Verifications that 1. and 2. above hold for $a(v,w) = \phi(v+w)-\phi(v)-\phi(w)$ follow; cases 3. and  4. also follow quite simply using the same substitutions for the quadratic form $\phi$ as used below.  

1. 
$$
\begin{aligned}
a(u+v,w) &= a(u,w) + a(v,w)\\
\phi(u+v+w)-\phi(u+v)-\phi(w) &= \phi(u+w)-\phi(u)-\phi(w) + \phi(v+w)-\phi(v)-\phi(w)\\
\phi(u+w) + \phi(w+v)-\phi(v)-\phi(u)-2\phi(w) &= \phi(u+w)+ \phi(v+w)-\phi(v)-\phi(u)-2\phi(w)\\
\end{aligned}
$$

=============== 

2. 
$$
\begin{aligned}
a(\lambda u,v ) = \lambda a(u,v)\\
\phi(\lambda u+v)-\phi( \lambda u)-\phi(v) &= \lambda \phi(u+v)-\lambda\phi(u)-\lambda\phi(v)\\
\lambda \phi(u+v) - \lambda \phi(u) - \lambda \phi(v) + \lambda^2 \phi(u) + \phi(v)- \lambda^2\phi( u)-\phi(v) &= \lambda \phi(u+v)-\lambda\phi(u)-\lambda\phi(v) \\
\lambda \phi(u+v) - \lambda \phi(u) - \lambda \phi(v) &= \lambda \phi(u+v)-\lambda\phi(u)-\lambda\phi(v)) \\
\end{aligned}
$$


## Question 2
Show that adding rows to a matrix (e.g. more data in a least squares problem) can only increase the smallest singular value.

## Answer 2

From the max-min theorem for singular values we can determine the value of the smallest singular value of a matrix $A \in \mathcal{C}^{m \times n}$:
$$
\sigma(A)_{min} = \min_{x \in \mathcal{C}^n \\ ||\mathbf{x}||_2 =1} ||\mathbf{A}\mathbf{x}||_2.
$$
Similarly, for the matrix $A_1 \in \mathcal{C}^{m +1  \times n}$ such that 
$$
A_1 = \begin{bmatrix}  \mathbf{A} \\ \mathbf{z} \end{bmatrix}
$$

we have the following

$$
\sigma(A_1)_{min} = \min_{x \in \mathcal{C}^n \\ ||\mathbf{x}||_2 =1} ||\mathbf{A_1}\mathbf{x}||_2.
$$

Observe that 
$$
\min_{x \in \mathcal{C}^n \\ ||\mathbf{x}||_2 =1}||\mathbf{A_1}\mathbf{x}||_2 = \min_{x \in \mathcal{C}^n \\ ||\mathbf{x}||_2 =1}||\mathbf{A}\mathbf{x} + \mathbf{z}\mathbf{x}||_2 \le \min_{x \in \mathcal{C}^n \\ ||\mathbf{x}||_2 =1} (||\mathbf{A}\mathbf{x}||_2 + ||\mathbf{z}\mathbf{x}||_2).
$$

Clearly, 
$$
||\mathbf{A}\mathbf{x}||_2  \le ||\mathbf{A}\mathbf{x}||_2 + ||\mathbf{z}\mathbf{x}||_2.
$$

Thus, 
$$
\sigma(A)_{min} \le \sigma(A_1)_{min}
$$
as desired. 

## Question 3

Write a code to minimize $$\sum_{j=0}^N |p(x_j)-\cos(x_j)|^2$$ where $x_j$ are points in a uniform mesh on $[-\pi, \pi]$, with $p(x) = c_0 + c_1 x^2 + c_2 x^4 + c_3 x^6$.

```julia
# Uniform mesh
num  = 100
vals = LinRange(-pi,pi,num)

# Matrix of coefficients
A    = zeros(num,4)
for j = 1:num
   A[j,:] = map(i -> vals[j].^i, [0,2,4,6])
end
y       = map(cos  , vals)

# Get min 
c = A\y

# Check sol
using Plots
plot(vals, cos)
poly(x) = c[1] + c[2]x^2 + c[3]x^4 + c[4]x^6
plot!(poly,vals)

```
